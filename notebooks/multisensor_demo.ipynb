```python
import numpy as np
import pandas as pd
import datetime as dt

# ----------------------------
# 1. CONFIG
# ----------------------------
np.random.seed(42)
duration_minutes = 180  # 3 hours of data
sampling_rate = 1       # 1 sample per second

n_samples = duration_minutes * 60

timestamps = [
    dt.datetime(2025, 1, 1, 9, 0, 0) + dt.timedelta(seconds=i)
    for i in range(n_samples)
]

# ----------------------------
# 2. GENERATE BASELINES
# ----------------------------
# HR baseline: 70-90 bpm
hr = np.random.normal(80, 5, n_samples)

# SpO baseline: 96-99%
spo2 = np.random.normal(98, 0.5, n_samples)

# Temperature baseline: 36.437.0C
temp = np.random.normal(36.6, 0.1, n_samples)

# Motion baseline: mostly low, occasional spikes
motion = np.random.exponential(scale=0.1, size=n_samples)

# ----------------------------
# 3. INJECT ANOMALIES
# ----------------------------
def inject_anomaly(signal, idx_start, idx_end, magnitude):
    signal[idx_start:idx_end] += magnitude

# Random anomaly windows
anom_windows = [
    (500, 560),     # HR spike
    (2500, 2600),   # Low SpO2 dip
    (4000, 4100),   # Temp elevation (fever-like)
    (6200, 6300),   # Big motion burst
]

# inject anomalies
inject_anomaly(hr,     500, 560,   25)     # HR spike
inject_anomaly(spo2,  2500, 2600,  -3)     # SpO2 drop
inject_anomaly(temp,  4000, 4100,   0.6)   # fever spike
inject_anomaly(motion,6200, 6300,   2.0)   # extreme movement

# ----------------------------
# 4. BUILD DATAFRAME
# ----------------------------
df = pd.DataFrame({
    "timestamp": timestamps,
    "heart_rate": hr,
    "spo2": spo2,
    "temperature": temp,
    "motion": motion
})

# ----------------------------
# 5. SAVE CSV
# ----------------------------
df.to_csv("sample_multisensor.csv", index=False)

df.head()
```

```python
from google.colab import files
files.download("sample_multisensor.csv")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("sample_multisensor.csv", parse_dates=["timestamp"])

# ---------------------------
# 1. NORMALIZATION (z-score)
# ---------------------------
signals = ["heart_rate", "spo2", "temperature", "motion"]

for s in signals:
    df[f"{s}_z"] = (df[s] - df[s].mean()) / df[s].std()

# ---------------------------
# 2. WEIGHTED FUSION
# ---------------------------
weights = {
    "heart_rate_z": 0.35,
    "spo2_z":       0.35,
    "temperature_z":0.15,
    "motion_z":     0.15
}

df["fusion_score"] = (
    df["heart_rate_z"]  * weights["heart_rate_z"] +
    df["spo2_z"]        * weights["spo2_z"] +
    df["temperature_z"] * weights["temperature_z"] +
    df["motion_z"]      * weights["motion_z"]
)

# ---------------------------
# 3. SMOOTHING (rolling window)
# ---------------------------
df["fusion_smooth"] = df["fusion_score"].rolling(60).mean()

# ---------------------------
# 4. ANOMALY FLAGGING
# ---------------------------
threshold = df["fusion_smooth"].mean() + 2 * df["fusion_smooth"].std()
df["anomaly"] = df["fusion_smooth"] > threshold

df.head()
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# -----------------------------
# Create sliding window dataset
# -----------------------------

class FusionDataset(Dataset):
    def __init__(self, df, window_size=60):
        self.window_size = window_size
        features = ["heart_rate_z", "spo2_z", "temperature_z", "motion_z"]
        self.data = df[features].values.astype("float32")

    def __len__(self):
        return len(self.data) - self.window_size

    def __getitem__(self, idx):
        window = self.data[idx : idx + self.window_size]
        return torch.tensor(window)

dataset = FusionDataset(df, window_size=60)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

len(dataset)
```

```python
import torch.nn as nn

class LSTMAutoencoder(nn.Module):
    def __init__(self, n_features=4, embedding_dim=16):
        super().__init__()

        self.encoder = nn.LSTM(
            input_size=n_features,
            hidden_size=embedding_dim,
            batch_first=True
        )

        self.decoder = nn.LSTM(
            input_size=embedding_dim,
            hidden_size=n_features,
            batch_first=True
        )

    def forward(self, x):
        # Encode
        encoded, _ = self.encoder(x)

        # Need only the final hidden sequence repeated
        last_step = encoded[:, -1, :].unsqueeze(1)
        repeated = last_step.repeat(1, x.size(1), 1)

        # Decode
        decoded, _ = self.decoder(repeated)
        return decoded

model = LSTMAutoencoder()
model
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

epochs = 10
loss_list = []

for epoch in range(epochs):
    epoch_loss = 0
    for batch in dataloader:
        batch = batch.to(device)
        optimizer.zero_grad()

        output = model(batch)
        loss = criterion(output, batch)
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    loss_list.append(epoch_loss / len(dataloader))
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss_list[-1]:.6f}")
```

```python
model.eval()

errors = []
with torch.no_grad():
    for i in range(len(dataset)):
        seq = dataset[i].unsqueeze(0).to(device)
        recon = model(seq)
        loss = criterion(recon, seq).item()
        errors.append(loss)

errors = np.array(errors)
threshold_lstm = errors.mean() + 2 * errors.std()

df_lstm = df.iloc[60:].copy()
df_lstm["recon_error"] = errors
df_lstm["lstm_anomaly"] = df_lstm["recon_error"] > threshold_lstm

df_lstm.head()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df_lstm["timestamp"], df_lstm["recon_error"], label="Reconstruction Error")
plt.scatter(
    df_lstm["timestamp"][df_lstm["lstm_anomaly"]],
    df_lstm["recon_error"][df_lstm["lstm_anomaly"]],
    color="red", s=10, label="Anomaly"
)
plt.axhline(threshold_lstm, color="orange", linestyle="--", label="Threshold")
plt.title("LSTM Fusion Model  Reconstruction Error & Anomalies")
plt.legend()
plt.tight_layout()
plt.show()
```
